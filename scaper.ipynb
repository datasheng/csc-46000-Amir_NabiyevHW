```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ccny.cuny.edu/registrar/fall'
```

```python
response = requests.get(url)
html_doc = response.text
soup = BeautifulSoup(html_doc)
pretty_soup = soup.prettify()
```

```python
text = soup.find_all('table')
table = text[0]
contents = table.find_all('tr')

table_dict = {
    'date': [],
    'date_of_week': [],
    'main_text': [],
}

for row in contents:
    cells = row.find_all('td')
    if len(cells) >= 3: # just to avoid that first table row 
        date_text = cells[0].get_text(strip = True)
        dow_text = cells[1].get_text(strip = True)
        main_text = cells[2].get_text(strip = True)
        table_dict['date'].append(date_text)
        table_dict['date_of_week'].append(dow_text)
        table_dict['main_text'] .append(main_text)
```

```python
df = pd.DataFrame(table_dict)
df.head(10)
```
